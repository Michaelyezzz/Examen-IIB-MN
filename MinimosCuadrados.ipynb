```python
from src import ajustar_min_cuadrados
# Derivadas parciales para regresión cuadrática
# ####################################################################
def der_parcial_2(xs: list, ys: list) -> tuple[float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro a (x²)"""
    c_2 = sum(xi * xi * xi * xi for xi in xs)  # coeficiente de a
    c_1 = sum(xi * xi * xi for xi in xs)      # coeficiente de b
    c_0 = sum(xi * xi for xi in xs)           # coeficiente de c
    c_ind = sum(yi * xi * xi for xi, yi in zip(xs, ys))  # término independiente
    return (c_2, c_1, c_0, c_ind)

def der_parcial_1(xs: list, ys: list) -> tuple[float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro b (x)"""
    c_2 = sum(xi * xi * xi for xi in xs)      # coeficiente de a
    c_1 = sum(xi * xi for xi in xs)           # coeficiente de b
    c_0 = sum(xi for xi in xs)                # coeficiente de c
    c_ind = sum(yi * xi for xi, yi in zip(xs, ys))  # término independiente
    return (c_2, c_1, c_0, c_ind)

def der_parcial_0(xs: list, ys: list) -> tuple[float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro c (término constante)"""
    c_2 = sum(xi * xi for xi in xs)           # coeficiente de a
    c_1 = sum(xi for xi in xs)                # coeficiente de b
    c_0 = len(xs)                             # coeficiente de c
    c_ind = sum(ys)                           # término independiente
    return (c_2, c_1, c_0, c_ind)
```

```python
# Ajustar los coeficientes para la ecuación cuadrática
a, b, c = ajustar_min_cuadrados(xs, ys, gradiente=[der_parcial_2, der_parcial_1, der_parcial_0])

# Generar puntos para la curva
x = np.linspace(min(xs), max(xs), 200)
y = [a * xi**2 + b * xi + c for xi in x]

print(f"Coeficientes de la parábola y = ax² + bx + c:")
print(f"a: {a:.6f}")
print(f"b: {b:.6f}")
print(f"c: {c:.6f}")

# Calcular errores relativos
errores_relativos = [abs((yi - (a * xi**2 + b * xi + c)) / yi) for xi, yi in zip(xs, ys)]
promedio_error_relativo = np.mean(errores_relativos)

# Encontrar el máximo error relativo y su ubicación
max_error = max(errores_relativos)
indice_max_error = errores_relativos.index(max_error)
x_max_error = xs[indice_max_error]
y_max_error = ys[indice_max_error]
y_ajustado = a * x_max_error**2 + b * x_max_error + c

print(f"\nEstadísticas de error:")
print(f"Promedio del error relativo: {promedio_error_relativo:.2e}")
print(f"Máximo error relativo: {max_error:.2e}")
print(f"Ubicación del máximo error:")
print(f"  x = {x_max_error}")
print(f"  y real = {y_max_error}")
print(f"  y ajustado = {y_ajustado:.4f}")
print(f"  Error absoluto = {abs(y_max_error - y_ajustado):.4f}")

# Visualizar resultados
plt.figure(figsize=(10, 6))
plt.scatter(xs, ys, color='blue', label='Datos', alpha=0.5)
plt.plot(x, y, color="red", label=r"$y = ax^2 + bx + c$")
plt.scatter([x_max_error], [y_max_error], color='green', s=100, 
           label='Máximo error', zorder=5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajuste cuadrático por mínimos cuadrados")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
